```python
from bs4 import BeautifulSoup 
import requests 
from csv import writer
```

### Get URL

```python
url = 'https://www.iproperty.com.my/sale/all-residential/?q=kuala%20lumpur'
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

### Get parent category where all the info are wrap inside

```python
lists = soup.find_all('div', class_ = "PremiumCardstyle__CardWrapper-cvkMVX jRRXhG")
```

```python
print(lists)
```

### Find all the details of the real estate posting

```python
for list in lists:
    number = list.find('span', class_ = "phone-number").text
    location = list.find('div', class_ = "PremiumCardstyle__AddressWrapper-ldsjqp gRJjrp").text
    price = list.find('div', class_ = "ListingPricestyle__ListingPriceWrapper-eZrVug gUGYlt listing-price").text
    description = list.find('p', class_ = "ListingAttributesstyle__ListingAttrsDescriptionItemWrapper-fQKuaA fJmpgN attributes-description-item").text
    info = [number, location, price, description]
    print(info)
```

### Replace anything if needed

```python
for list in lists:
    number = list.find('span', class_ = "phone-number").text.replace('\xa0•\xa0 ', '-')
    location = list.find('div', class_ = "PremiumCardstyle__AddressWrapper-ldsjqp gRJjrp").text.replace('\xa0•\xa0 ', '-')
    price = list.find('div', class_ = "ListingPricestyle__ListingPriceWrapper-eZrVug gUGYlt listing-price").text.replace('\xa0•\xa0 ', '-')
    description = list.find('p', class_ = "ListingAttributesstyle__ListingAttrsDescriptionItemWrapper-fQKuaA fJmpgN attributes-description-item").text.replace('\xa0•\xa0 ', '-')
    info = [number, location, price, description]
    print(info)
```

### Import to csv file 

```python
with open('house.csv', 'w', encoding='utf8', newline='') as file:
    thewriter = writer(file)
    header = ["Number", "Location", "Price", "Description"]
    thewriter.writerow(header)


    for list in lists:
        number = list.find('span', class_ = "phone-number").text.replace('\xa0•\xa0 ', '-')
        location = list.find('div', class_ = "PremiumCardstyle__AddressWrapper-ldsjqp gRJjrp").text.replace('\xa0•\xa0 ', '-')
        price = list.find('div', class_ = "ListingPricestyle__ListingPriceWrapper-eZrVug gUGYlt listing-price").text.replace('\xa0•\xa0 ', '-')
        description = list.find('p', class_ = "ListingAttributesstyle__ListingAttrsDescriptionItemWrapper-fQKuaA fJmpgN attributes-description-item").text.replace('\xa0•\xa0 ', '-')
        info = [number, location, price, description]
        thewriter.writerow(info)
```
